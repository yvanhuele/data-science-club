```python
from scipy import stats
import matplotlib.pyplot as plt
import numpy as np
from copy import copy
%matplotlib inline
```

```python
class bandit:
    def __init__(self, mus = None, d = None):
        if d is None:
            if not mus is None:
                print 'mus:', mus
                d = len(mus)
            elif d is None:
                d = 10
        if mus is None:
            mu_dist = stats.uniform(0,1)
            mus = mu_dist.rvs(d)
        self.mus = mus
    
    def pull(self, j):
        return stats.norm(self.mus[j], .5).rvs(1)[0]
```

```python
N = 1000
d = 10


mus = [.3, .3, .2, .2, .2, .2, .2, .8, .8, 1.0]
mus = sorted(mus)
game = bandit(d = d, mus = mus)
#Q = [0] * d
total_rewards = [0] * d
n_played = [1] * d
arms_pulled = []
rewards = []

n_played_list = []

for i in range(1,N):
    avg_rewards = [total_rewards[j] / float(n_played[j]) for j in range(d)  ]
    unexplored_bonus = [np.sqrt( 2 * np.log(i) / n_played[j]) for j in range(d)  ]
    score = [avg_rewards[j] + unexplored_bonus[j] for j in range(d)  ]
    #print 'Q:', Q
    arm = score.index(max(score))
    reward = game.pull(arm)
    print 'arm:', arm, 'reward:', reward
    arms_pulled.append(arm)
    n_played[arm] += 1
    n_played_list.append(np.array(copy(n_played)) / float(sum(n_played))  )
    rewards.append(reward)
    total_rewards[arm] += reward
    
    
```

```python

x = range(N-1)
A = np.array(n_played_list)
plt.clf()
fig, ax = plt.subplots()
#ax.stackplot(range(N-1), A)

labels = [np.round(mu,2) for mu in game.mus]


ax.stackplot(np.linspace(0,1,N-1), A.T, labels = labels)

ax.legend(loc=3)
plt.xlim(-.3, 1.2)
plt.show()
```
